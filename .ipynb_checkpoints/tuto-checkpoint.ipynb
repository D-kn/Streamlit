```python
import streamlit as st
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'col1': [i for i in range(1, 5)],
    'col2': [i for i in range(10, 41, 10)]
})

df
```

```python
dataframe = np.random.randn(10, 20)
st.dataframe(dataframe)
```

## Draw a line chart


```python
chart_data = pd.DataFrame(
    np.random.randn(20, 3),
    columns = ['a', 'b', 'c']
)    
st.line_chart(chart_data)
```
